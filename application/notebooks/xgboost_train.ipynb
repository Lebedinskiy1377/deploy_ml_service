```python
import xgboost as xgb
import optuna
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error
from sklearn.preprocessing import LabelEncoder


X = df.drop(['dates', 'price_per_sku', 'creation_date', 'expiration_date'], axis=1)
y = df.price_per_sku

cat_str_features = ['fincode', 'ui1_code', 'ui2_code', 'ui3_code', 'vendor', 'brand_code']

cat_features = ['year', 'month', 'week_num', 'day']
X_xgb = X.copy()


for col in cat_str_features:
    le = LabelEncoder()
    X_xgb[col] = le.fit_transform(X_xgb[col])
    X_xgb[col] = X_xgb[col].astype('category')

for col in cat_features:
    X_xgb[col] = X_xgb[col].astype('category')

def objective(trial):
    param = {
        'objective': 'reg:squarederror',
        'booster': 'gbtree',
        'eval_metric': 'rmse',
        'verbosity': 0,
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'lambda': trial.suggest_float('lambda', 0.1, 2.0),
        'alpha': trial.suggest_float('alpha', 0.1, 2.0),
        'min_child_weight': trial.suggest_float('min_child_weight', 1, 10),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'gamma': trial.suggest_float('gamma', 0, 1),
        'random_state': RANDOM_STATE
    }
    
    tss = TimeSeriesSplit(10)
    rmse_list = []
    
    for train_index, val_index in tss.split(X_xgb, y):
        X_train, X_val = X_xgb.iloc[train_index], X_xgb.iloc[val_index]
        y_train, y_val = y.iloc[train_index], y.iloc[val_index]
        
        dtrain = xgb.DMatrix(X_train, label=y_train, enable_categorical=True)
        dval = xgb.DMatrix(X_val, label=y_val, enable_categorical=True)
        
        xgb_regressor = xgb.train(param, dtrain, verbose_eval=False)
        y_pred = xgb_regressor.predict(dval)
        
        rmse_list.append(np.sqrt(mean_squared_error(y_val, y_pred)))
        
    return np.mean(rmse_list)

study_xgb = optuna.create_study(direction='minimize')
study_xgb.optimize(objective, n_trials=10)
```

```python
X_train_xgb = df.drop(['dates', 'price_per_sku', 'creation_date', 'expiration_date'], axis=1)
y = df.price_per_sku

cat_features = ['year', 'month', 'week_num', 'day']

for col in cat_features:
    X_train_xgb[col] = X_train_xgb[col].astype('category')
    
cat_str_features = ['fincode', 'ui1_code', 'ui2_code', 'ui3_code', 'vendor', 'brand_code']

for col in cat_str_features:
    le = LabelEncoder()
    X_train_xgb[col] = le.fit_transform(X_train_xgb[col])
    X_train_xgb[col] = X_train_xgb[col].astype('category')

dtrain = xgb.DMatrix(X_train_xgb, label=y, enable_categorical=True)

xgb_regressor = xgb.train(study_xgb.best_params, dtrain)

X_inf = test.drop(['dates', 'creation_date', 'expiration_date'], axis=1)
X_inf = X_inf[X_train_xgb.columns]
for col in cat_features:
    X_inf[col] = X_inf[col].astype('category')
    
for col in cat_str_features:
    le = LabelEncoder()
    X_inf[col] = le.fit_transform(X_inf[col])
    X_inf[col] = X_inf[col].astype('category')

dinf = xgb.DMatrix(X_inf, enable_categorical=True)

y_pred = xgb_regressor.predict(dinf)
```
